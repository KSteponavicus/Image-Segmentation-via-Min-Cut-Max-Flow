```python
class Graph:

    def __init__(self, num_pixels: int):
        self.total_nodes = num_pixels + 2 # pixels + source node + sink
        self.source = num_pixels # ID for source
        self.sink = num_pixels + 1 # ID for sink

        self.adj_list = [[] for _ in range(self.total_nodes)] 

    def add_edge(self, u: int, v:int, capacity:float):
        
        u_index = len(self.adj_list[u])
        v_index = len(self.adj_list[v])

        forward_edge = Edge(link_to = v, capacity=capacity, rev_index = v_index)
        reverse_edge = Edge(to=u, capacity=0.0, rev_index=u_index)

        # Push them into the adjacency lists
        self.adj_list[u].append(forward_edge)
        self.adj_list[v].append(reverse_edge)
```

```python
#BFS 

from collections import deque

def BFS_residual(graph, source, sink):

    # Track paths: maps child_node_id -> (parent_node_id, edge_object used)
    parent = {v: None for v in graph.total_nodes}

    # Simple visited set to prevent infinite loops
    visited = set()

    queue = deque([source])
    visited.add(source)

    while queue:
        u = queue.popleft()

        if  u == sink:
            return parent
        
        for edge in graph.adj_list[u]:
            v = edge.link_to

            if v not in visited and edge.residual_capacity() > 0:
                visited.add(v)
                parent[v] = (u, edge)
                queue.append(v)

    # Track paths: maps child_node_id -> (parent_node_id, edge_object used) 
    return None 
```
